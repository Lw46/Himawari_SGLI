```python
import cdsapi

def month_size(month,YYYY):
    big_month=['01','03','05','07','08','10','12']
    small_month=['04','06','09','11']
    if month in big_month:
        s = 31
    elif month in small_month:
        s = 30
    elif month == '02':
        if (int(YYYY) % 4 == 0 and int(YYYY) % 100 != 0) or int(YYYY) % 400 == 0:
            s = 29
        else:
            s = 28
    return s

def download_atmos (Y,M,H) :
    D = month_size(M,Y)
    c = cdsapi.Client()
    c.retrieve(
    'cams-global-reanalysis-eac4',
    {
        'variable': [
            'total_column_ozone', 'total_column_water_vapour',
        ],
        'format': 'netcdf',
        'time': '{}:00'.format(H),
        'date': '{}-{}-01/{}-{}-{}'.format(Y,M,Y,M,D),
    },
    '/media/liwei/Data/CAMS/{}{}{}.nc'.format(Y,M,H))
```

```python
YYYY = ['2015','2016','2017','2018','2019','2020','2021']
MM = ['01','02','03','04','05','06','07','08','09','10','11','12']
MIN = ['20','30']
HH = ['00','03','06','09','12','15','18','21']
for y in YYYY:
    for m in MM:
        for h in HH:
            download_atmos (y,m,h)
```
