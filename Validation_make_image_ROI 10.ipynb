```python
def identifer(a):
    mean = np.nanmean(a)
    std = np.nanstd(a) 
    down = mean - 3 * std
    up =  mean + 3 * std
    n_a = np.where(a < down,np.nan,a)
    n_a = np.where(n_a > up,np.nan,n_a)
    return n_a
```
